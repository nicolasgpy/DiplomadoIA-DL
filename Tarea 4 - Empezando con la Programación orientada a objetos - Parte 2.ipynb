# Punto 2
Escriba una clase iterable que tenga la siguiente funcionalidad

- ## Punto 2.1:Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.

- ## Punto 2.2: Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
- ## Punto 2.3: Entregue un error si los datos se han agotado

```python
class Get_df_iris(object):
  def __init__(self, df):
    self.dataframe = pd.read_csv(df, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
    self.limit = len(self.dataframe)
    self.val = 2
    self.val_ini = 0

  def __iter__(self):
    return self
  
  def __next__(self):
    if self.val > self.limit:
      raise Exception("Sorry, no more rows")
    else:
      return_val = self.dataframe[self.val_ini:self.val]
      self.val_ini += 2
      self.val += 2
      return return_val
```

```python
for i in Get_df_iris(url):
  print('-------------------------------------------')
  print(i, end=', ')
```
